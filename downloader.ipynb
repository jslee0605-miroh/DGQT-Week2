```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
```

```python
#TREASURIES TICKERS
#tickers = ["ZB=F", "ZN=F", "ZF=F", "ZT=F"] 
tickers = ["BZ=F", "NG=F", "CL=F", "YM=F", "URA"]
#brent, nat gas, wti, YM=F is dow jones industrial, URA is Uranium

for ticker in tickers:
    data = yf.download(ticker, period="60y", interval="1d")

    # Save S&P 500 data to CSV
    data.to_csv(f"{ticker}_daily_12years.csv", header=True) 
    print("f{ticker} data saved")
```
